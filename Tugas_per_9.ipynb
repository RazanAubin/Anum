```python
import numpy as np
import matplotlib.pyplot as plt

# Pendefinisian fungsi
def f(x):
    a = 0.2
    b = 1.4
    c = -8
    d = -23
    return a*x**3 + b*x**2 + c*x + d

# Pengaturan parameter
x_awal = 8
x_akir = -12

# Visualisasi
x = np.linspace(x_awal, x_akir, 10000)
y = f(x)

plt.figure(figsize=(6, 6.5))
plt.plot(x, y, '-r', label='y')

# Menemukan titik-titik di mana nilai fungsi sama dengan 0
zero_crossings = np.where(np.diff(np.sign(y)))[0]
zero_crossings_x = x[zero_crossings]
zero_crossings_y = y[zero_crossings]

plt.plot(zero_crossings_x, zero_crossings_y, 'go')  # Menambahkan titik-titik
plt.plot([x_awal, x_akir], [0, 0], '-k')  # Menambahkan sumbu-x
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Pendefinisian fungsi
def f(x):
    a = 1
    b = -2
    c = -11
    d = 12
    return a*x**3 + b*x**2 + c*x + d

# Turunan pertama dari fungsi
def df(x):
    return 3*0.2*x**2 + 2*1.4*x - 8

# -- Metode False Position --
def false_position(xl, xu, err_m):
    xr = xu - (f(xu) * (xu-xl)) / (f(xu) - f(xl))
    while abs(f(xr)) > err_m:
        if f(xl)*f(xr) < 0: # Menentukan jika lokasi berada diantara f(xl) dan f(xr)
            xu = xr
        else:
            xl = xr
        xr = xu - (f(xu) * (xu-xl)) / (f(xu) - f(xl))
    return xr

# -- Penyelesaian --
# Pengaturan parameter
x_awal = float(input("Masukkan nilai x_awal: "))
x_akhir = float(input("Masukkan nilai x_akhir: "))
err_m = float(input("Masukkan nilai err_m: "))
dx = float(input("Masukkan nilai dx: "))

# Metode False Position untuk mencari akar
roots_false_position = []
x = np.arange(x_awal, x_akhir, dx)
for i in range(len(x)-1):
    if f(x[i])*f(x[i+1]) < 0:
        root = false_position(x[i], x[i+1], err_m)
        roots_false_position.append(root)

# -- Visualisasi --
x = np.linspace(x_awal, x_akhir, 10000)
plt.figure(figsize=(6, 6.5))
sumbu_x = x - x - 0
y = f(x)
plt.plot(x, sumbu_x, '-k')
plt.plot(x, y, '-r', label='y')

# Menambahkan scatter plot untuk akar
plt.scatter(roots_false_position, [f(root) for root in roots_false_position], color='green', label='Akar')

plt.legend(loc='upper left')
plt.grid()
plt.show()

# Menampilkan akar-akar fungsi dalam interval menggunakan metode false position
print("\nAkar-akar fungsi dalam interval menggunakan metode false position dari", x_awal, "sampai", x_akhir, "adalah:")
for root in roots_false_position:
    print("x =", root, ", y =", f(root))
```

```python
import matplotlib.pyplot as plt

a = int(input("a: "))
b = int(input("b: "))
c = int(input("c: "))
d = int(input("d: "))

x_awal = -4
x_akhir = 5
N = 1000000

dx = (x_akhir - x_awal) / N
x = x_awal
luas = 0
iterasi = 0

def f_x(a, b, c, d, x):
    return a*x**3 + b*x**2 + c*x + d

x_points = []
y_points = []

while x <= x_akhir:
    luas_keping = f_x(a, b, c, d, x) * dx
    luas += luas_keping
    x += dx
    iterasi += 1
    if iterasi % 1000 == 0:
        x_points.append(x)
        y_points.append(f_x(a, b, c, d, x))

        print(f"Iterasi: {iterasi}")
        print(f"Luas keping = {luas_keping}")
        print(f"x = {x}")
        print(f"Luas total = {luas}")

plt.plot(x_points, y_points)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Kurva f(x)')
plt.grid(True)
plt.show()
```
